### Regrecion Lineal con Redes Neuronales

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
#definir un dato de entrenamiento
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# y= 2x - 1         y=mx+b
```

```python
#definir nuestra red neuronal
model = tf.keras.Sequential([keras.layers.Dense(1, input_shape=[1]),
                             keras.layers.Dense(10),
                             keras.layers.Dense(1),
                            ])
```

```python
#sgd = desenso del gradiente estocatico
#mean squared error = error cuadratico medio
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
history = model.fit(xs,ys,epochs=1000)
```

```python
#predicciones
result = model.predict([6])
result
```

```python
plt.xlabel("Epoch numbers")
plt.ylabel("Loos Magnitude")
plt.plot(history.history["loss"])
```

### Convecion de celsius a farenheint

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
#definir un dato de entrenamiento
cel = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
far = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

```python
#Capa
l1 =keras.layers.Dense(1, input_shape=[1])
l2 =keras.layers.Dense(10)
l3 =keras.layers.Dense(20)
l4 =keras.layers.Dense(1)
model = tf.keras.Sequential([l1,l2,l3,l4])
```

```python
model.compile(optimizer=keras.optimizers.Adam(0.1), loss="mean_squared_error")
```

```python
history = model.fit(cel,far,epochs=500,verbose=False)
print("Entrenamiento finalisado")
```

```python
#predicciones
print(model.predict([-40]))
```

```python
print("valores de la capa l1, neurona {}= ".format(l1.get_weights()))
```

```python
plt.xlabel("Epoch numbers")
plt.ylabel("Loos Magnitude")
plt.plot(history.history["loss"])
```
